DATASET CLEANER 

```python
import pandas as pd
```

```python
# Path to your raw CSV
input_file = r"C:\F1_analysis\data\raw\laps_2024.csv"
output_file = r"C:\F1_analysis\data\processed\laps_2024_cleaned.csv"
```

```python
# Load the data
df = pd.read_csv(input_file, parse_dates=['LapTime'])
```

```python
df.columns = df.columns.str.strip()
```

```python
# Convert LapTime to timedelta
# ---------------------------
df['LapTime'] = pd.to_timedelta(df['LapTime'])
```

```python
# Filter invalid laps
# ---------------------------
# Start with LapTime not NaN
valid_mask = df['LapTime'].notna()

# Filter Deleted if column exists
if 'Deleted' in df.columns:
    # Ensure it's boolean
    if df['Deleted'].dtype != bool:
        df['Deleted'] = df['Deleted'].astype(bool)
    valid_mask &= df['Deleted'] == False

df = df[valid_mask]
```

```python
# Add LapTimeSeconds
# ---------------------------
df['LapTimeSeconds'] = df['LapTime'].dt.total_seconds()
```

```python
df.head()
```

```python
# Drop unnecessary columns
# ---------------------------
drop_cols = [
    'Sector1Time','Sector2Time','Sector3Time',
    'Sector1SessionTime','Sector2SessionTime','Sector3SessionTime',
    'SpeedI1','SpeedI2','SpeedFL','SpeedST',
    'LapStartTime','LapStartDate','TrackStatus','Position',
    'Deleted','DeletedReason','FastF1Generated','IsAccurate',
    'Season','Date'
]
```

```python
drop_cols = [col for col in drop_cols if col in df.columns]
df = df.drop(columns=drop_cols)
```

```python
# Save cleaned CSV
# ---------------------------
df.to_csv(output_file, index=False)
print(f"✅ Cleaned lap data saved to {output_file}")
```
