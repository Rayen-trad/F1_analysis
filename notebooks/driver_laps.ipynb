```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("C:\F1_analysis\data\processed\laps_2024_cleaned.csv")
```

```python
#df.info()
#df.describe()
df.head()
```

```python
driver = "VER"
race_name = "Bahrain Grand Prix"

driver_laps = df[
    (df["Driver"] == driver) &
    (df["Race_Name"] == race_name)
].copy()
```

```python
COMPOUND_COLORS = {
    "SOFT": "#FF3333",        # Red
    "MEDIUM": "#FFD700",      # Yellow
    "HARD": "#111111",        # White
    "INTERMEDIATE": "#39B54A",# Green
    "WET": "#1E90FF"          # Blue
}
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=driver_laps,
    x="LapNumber",
    y="LapTimeSeconds",
    hue="Compound",
    palette=COMPOUND_COLORS,
    s=80,
    linewidth=0
)

plt.gca().invert_yaxis()
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (seconds)")
plt.title(f"{driver} Lap Times  {race_name}")
plt.grid(True)
plt.legend(title="Tyre Compound")

plt.tight_layout()
plt.show()
```
